```python
import pickle
import pandas as pd
from tqdm import tqdm
```

```python
files_list = os.listdir('LA_extracted_features/')
```

```python
all_features = []
for file in tqdm(files_list):
  data=pickle.load(open(f'LA_extracted_features/{file}','rb'))
  # print(len(data))
  all_features.extend(data)
```

```python
pd.DataFrame(all_features).to_csv('image_features.csv',index=False)
```
