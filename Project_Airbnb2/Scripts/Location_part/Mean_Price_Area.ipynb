```python
import pandas as pd
pd.set_option('max_columns', 200)
from google.colab import drive 
drive.mount('/content/drive')
```

```python
data = pd.read_pickle('/content/drive/MyDrive/Colab Notebooks/Capstone/data_zip.pkl')
data["price"] = data.price.str.replace("$","")
data["price"] = data.price.str.replace(",","")

data["price"] = data["price"].astype('float32')
```

```python

data_for_cal = data[["id","latitude","longitude","accommodates","beds","price"]]
```

```python
def average_cal(ac,lat,lon, data_for_cal):
  data_temp = data_for_cal[data_for_cal["accommodates"]==ac]
  data_temp = data_temp[(data_temp["latitude"] > lat - 0.01) & (data_temp["latitude"] < lat + 0.01) ]
  data_temp = data_temp[(data_temp["longitude"] > lon - 0.01) & (data_temp["longitude"] < lon + 0.01) ]

  return data_temp["price"].mean()
```

```python
data_for_cal["mean_area_accommodates_price"] = data_for_cal.apply(lambda x: average_cal(x["accommodates"], x["latitude"], x["longitude"], data_for_cal), axis=1)
```

```python
def average_cal_beds(beds,lat,lon, data_for_cal):
  data_temp = data_for_cal[data_for_cal["beds"]==beds]
  data_temp = data_temp[(data_temp["latitude"] > lat - 0.01) & (data_temp["latitude"] < lat + 0.01) ]
  data_temp = data_temp[(data_temp["longitude"] > lon - 0.01) & (data_temp["longitude"] < lon + 0.01) ]

  return data_temp["price"].mean()
```

```python
data_for_cal["mean_area_beds_price"] = data_for_cal.apply(lambda x: average_cal_beds(x["beds"], x["latitude"], x["longitude"], data_for_cal), axis=1)
```
