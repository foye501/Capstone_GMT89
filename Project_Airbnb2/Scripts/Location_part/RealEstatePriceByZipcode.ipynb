```python
make df of a list of Airbnb
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Capstone/listings.csv.gz', compression='gzip',
                   error_bad_lines=False)


data["price"] = data.price.str.replace("$","")
data["price"] = data.price.str.replace(",","")

data["price"] = data["price"].astype('float32')

```

```python

# to get zipzode
from uszipcode import SearchEngine
search = SearchEngine()


def get_zipcode(lat, lon):
    result = search.by_coordinates(lat = lat, lng = lon, returns = 1)
    try :
      return result[0].zipcode
    except IndexError:
      return None

data['zipcode'] = data.apply(lambda x: get_zipcode(x.latitude,x.longitude), axis=1)

pd.to_pickle(data, '/content/drive/MyDrive/Colab Notebooks/Capstone/data_zip.pkl')

```
