# Explaining Columns



*   supermarket
> Number of supermarkets within a 1 km radius of latitude and longitude


*   BusStop_distance
> The distance to the nearest bus stop within a radius of 500 meters from the entered latitude and longitude.The units of distance are not meters or miles, it is the difference of latitude and longitude obtained from the theorem of three squares,. If no bus stop is found within 500m, it was entered in np.inf.



*   Station_distance
> The distance to the nearest train and metro station within a radius of 500 meters from the entered latitude and longitude.The units of distance are not meters or miles, it is the difference of latitude and longitude obtained from the theorem of three squares,. If no station is found within 500m, it was entered in np.inf.


*   restaurant
> Number of restaurants within a 1 km radius of latitude and longitude



*   mean_area_accommodates_price
> Mean() the price in the Airbnb listing data, which is the same accommodations, in the area plus or minus 0.01 from the input latitude and longitude (0.01 is assumed to be approximately 1 km).


*   mean_area_beds_price
> Mean() the price in the Airbnb listing data, which is the same beds, in the area plus or minus 0.01 from the input latitude and longitude (0.01 is assumed to be approximately 1 km).




*   real_estate
> Average residential retail value calculated by zip code(from [Redfin](https://www.redfin.com/news/data-center/))
